# Visualisasi

## Line Plot

**Question** : Bagaimanakah Jumlah kasus di City of London dari tahun 2008-2016?


```python
data_tahuncl = pd.DataFrame(
               data[data["Wilayah"]== "City of London"].groupby(["Tahun"])["Value"].sum()
               ).sort_values(by='Tahun', ascending=True)
data_tahuncl
```

```python
data_tahuncl.plot(kind='line', figsize=(10,6))

plt.title('Total value City of London dari tahun 2008-2016')
plt.ylabel('Value')
plt.xlabel('Year')

plt.show()
```

**Analysis :** Dari plot diatas, Terlihat bahwa dari tahun 2008-2009 total kasus adalah 0, lalu terjadi peningkatan jumlah kasus pada tahun 2011-2012, dan kembali menurun pada tahun 20013. Pada tahun 2014-2016 mengalami kenaikan jumlah kasus.

## Histogram

**Question 1 :** Bagaimana jumlah kasus pada Westminster dari tahun 2008-2016?

```python
data_hist = pd.DataFrame(
            data[data["Wilayah"]== "Westminster"].groupby(["Tahun"])["Value"].sum()
            ).sort_values(by='Tahun', ascending=True)
data_hist.reset_index(inplace=True)
data_hist
```

```python
data_croydon = []
for i in range(len(data_hist)):
    for j in range(int(data_hist.Value[i])):
        data_croydon.append(data_hist.Tahun[i])
```

```python
plt.hist(data_croydon,bins=6, color='#b0db43')

plt.title('Histogram Total value Westminster tahun 2008-2016 ')
plt.show()
```

**Analysis :** Pada plot diatas, terlihat bahwa jumlah kasus di Westminster pada tahun 2008 cukup tinggi. Sedangkan pada tahun 2009-2010 mengalami penurunan. Sedangkan pada tahun 2011-2013 mengalami kenaikan jumlah kasus kemudian mengalami penrunan kembali di tahun 2014-2015. Pada tahun 2016 mengalami kenaikan kembali.

**Question 2:** Bagaimanakah Jumlah value Havering dan Barnet dari tahun 2008-2016?

```python
data_hist1 = pd.DataFrame(
            data[data["Wilayah"]== "Havering"].groupby(["Tahun"])["Value"].sum()
            ).sort_values(by='Tahun', ascending=True)
data_hist1 =data_hist1.rename(columns={"Value":"Havering"})
```

```python
data_hist2 = pd.DataFrame(
            data[data["Wilayah"]== "Barnet"].groupby(["Tahun"])["Value"].sum()
            ).sort_values(by='Tahun', ascending=True)
data_hist2 = data_hist2.rename(columns={"Value":"Barnet"})
```

```python
data_totalhist = pd.concat([data_hist1, data_hist2], axis=1, join="inner")
data_totalhist
```

```python
data_totalhist.plot(kind='hist', bins=10, figsize=(10,6), color=['#b0db43','#db2763'])

plt.title('Histogram Value Richmond upon Thames dan Merton pada tahun 2008-2016')
plt.ylabel('Years')
plt.xlabel('value')

plt.show()
```

**Question 3:**  5 Wilayah dengan total value tertinggi 3 tahun terakhir?

```python
datasatu = pd.DataFrame(
           data[data["Tahun"]== 2016].groupby(["Wilayah"])["Value"].sum()
           ).sort_values(by='Value', ascending=True)
datasatu = datasatu.tail(10)
datasatu = datasatu.rename(columns={"Value":"2016"})
```

```python
datadua = pd.DataFrame(
          data[data["Tahun"]== 2015].groupby(["Wilayah"])["Value"].sum()
          ).sort_values(by='Value', ascending=True)
datadua = datadua.tail(10)
datadua = datadua.rename(columns={"Value":"2015"})
```

```python
datatiga = pd.DataFrame(
           data[data["Tahun"]== 2014].groupby(["Wilayah"])["Value"].sum()
           ).sort_values(by='Value', ascending=True)
datatiga = datatiga.tail(10)
datatiga = datatiga.rename(columns={"Value":"2014"})
```

```python
data_top10 = pd.concat([datadua, datasatu], axis=1, join="inner")
```

```python
top_5 = pd.concat([datatiga, data_top10], axis=1)
top_5 = top_5.tail(5)
top_5
```

```python
top_5.plot(kind='barh', figsize=(12,6) , color=['#db2763','#b0db43','#ffc2de'])

plt.xlabel('Value')
plt.ylabel('Borough')
plt.title('Jumlah Total value 5 wilayah terbanyak dari tahun 2014-2016')
```

**Analysis:** Pada plot diatas, terlihat bahwa adanya penurunan pada Westminster tahun 2014-2016. Pada Camden terlihat adanya kenaikan dari tahun 2014 ke tahun 2015, lalu menurun pada tahun 2016. Pada Lambeth dan Newham terlihat adanya kenaikan tahun 2014-2016. Sedangkan pada Hillingdon terjadi penurunan pada tahun 2015, lalu kembali naik pada tahun 2016.

## Area Plot

**Question :** Jumlah Total value Wstrminster,Lambeth, dan Southwark pada tahun 2016 

```python
data['Tahun'] = data['Tahun'].astype('str')
```

```python
# memilih data wilayah westminster di tahun 2016
data_area1 = data[(data['Tahun']=='2016') & (data['Wilayah']=='Westminster')]

# mengelompokkan data dengan month 
data_area1 = (data_area1.groupby('Bulan')['Value']
                 .sum()
                 .reset_index()
                 .sort_values(by='Bulan', ascending=True)
                 .head(12))
data_area1 = pd.DataFrame(data_area1)
data_area1.head(12)
```

```python
# memilih data wilayah westminster di tahun 2016
data_area2 = data[(data['Tahun']=='2016') & (data['Wilayah']=='Lambeth')]

# mengelompokkan data dengan
data_area2  = (data_area2 .groupby('Bulan')['Value']
                 .sum()
                 .reset_index()
                 .sort_values(by='Bulan', ascending=True)
                 .head(12))
data_area2  = pd.DataFrame(data_area2 )
data_area2 .head(12)
```

```python
# memilih data wilayah Southwark di tahun 2016
data_area3  = data[(data['Tahun']=='2016') & (data['Wilayah']=='Southwark')]

# mengelompokkan data dengan
data_area3 = (data_area3.groupby('Bulan')['Value']
                 .sum()
                 .reset_index()
                 .sort_values(by='Bulan', ascending=True)
                 .head(12))
data_area3 = pd.DataFrame(data_area3)
data_area3.head(12)
```

```python
merged_df = pd.merge(data_area1, data_area2, on="Bulan")
df = pd.merge(merged_df, data_area3, on="Bulan")
df = df.rename(columns={'Value_x': 'Westminster', 'Value_y': 'Lambeth', 'Value': 'Southwark'}) 
df.set_index('Bulan')
```

```python
df.plot(kind='area', 
        color=['#b4da55','#fcf9f4','#f42069'] , #,'coolwarm_r', #'tab20_r',
        stacked=False,
        alpha =.8,
        x='Bulan',
        figsize=(12,6)# pass a tuple (x,y) size
        )
             
plt.title("Jumlah Total value Wstrminster,Lambeth, dan Southwark pada tahun 2016", fontsize=14)
plt.ylabel("Value")
plt.xlabel("Bulan")

plt.show()
```

**Analysis:** Pada Plot diatas, terlihat bahwa pada tahun 2016 wilayah Westminster terjadi kenaikan pada bulan ke-12. Pada wilayah lambeth terjadi kenaikan pada bulan ke-7. dan pada wilayah Southwark terjadi kenaikan pada bulan ke-10. 

## Pie Chart

**Question1 :** Bagiamanakah Jumlah value dari tahun 2008-2016?

```python
data_tahun = (data.groupby('Tahun')['Value']
                 .sum()
                 .reset_index()
                 .sort_values(by='Tahun', ascending=False)
                 .head(12))
data_tahun   = pd.DataFrame(data_tahun )
data_tahun
```

```python

colors = ["#7a0037","#bd457b", "#ff8abf", "#ffc2de", "#fcf9f4" ,"#f0fcda" ,"#e4febf", "#b4da55", "#7e8e01" ]
plt.figure(figsize=(8,6)) 

plt.pie(
    data_tahun['Value'],
    shadow=False,
    colors=colors,
    explode=(0.15, 0, 0, 0,0, 0 , 0,0 , 0),
    startangle=90,
    autopct='%1.1f%%',
    )


plt.axis('equal')
plt.legend(labels=data_tahun['Tahun'], loc="center left")


# View the plot
plt.title('Pie Chart Total Value Pertahun')
plt.tight_layout()
plt.show()
```

**Analysis :** Pada plot diatas, terlihat bahwa tahun tertinggi yaitu tahun 2012 dengan 12.0%, lalu tahun 2011 dengan 11.4%, lalu yang ketiga yaitu 2016 dengan 11.2%. Sedangkan presentasi terendah yaitu pada tahun 2014.

```python
data
```

```python
data_mayor = (data.groupby('Kejahatan Mayor')['Value']
                 .sum()
                 .reset_index()
                 .sort_values(by='Kejahatan Mayor', ascending=False)
                 .head(12))
data_mayor   = pd.DataFrame(data_mayor)
data_mayor
```

## Word Cloud

**Question:** Bagaiamanakah Wordcloud Wilayah dengan jumlah total value dari tahun 2008-2016?

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
values_Wilayah = data.Wilayah.unique()
words = {}
for value in values_Wilayah:
    words[value] = data[data.Wilayah ==value]['Value'].sum()
words
```

## Regression Plot

**Question:** Bagaimanakah Total value Southwark per tahun?

```python
import seaborn as sns
print("Seaborn installed and imported")
```

```python
data.head()
```

```python
data_so = (data['Tahun']=='2016') & (data['Wilayah']=='Southwark')
```

```python
data_so = pd.DataFrame(data.groupby('Tahun')['Value']
                 .sum()
                 .reset_index()
                 .sort_values(by='Tahun', ascending=True))
data_so   = data_so.set_index('Tahun')
```

```python
data_so.index = map(float, data_so.index)
```

```python
data_so.reset_index(inplace=True)
```

```python
data_so
```

```python
data_so.columns=['year','value']
```

```python
plt.figure(figsize=(15,10))

sns.set_style('whitegrid')

ax = sns.regplot(x='year', y='value', data=data_so)
ax.set(xlabel='Year', ylabel='Total Immigration') # add x- and y- labels
ax.set_title('Total Immigration to Canada from 1980-2013') #add title
```

**Analysis:** Pada plot diatas terlihat bahwa terjadi kenaikan dan penurunan value, lalu pada tahun 2014 terjadi kenaikan yg cukup siginfikan dari tahun 2014- 2016
