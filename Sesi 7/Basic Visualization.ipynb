# Area  Plots

Dengan sedikit modifikasi pada kode, kita dapat memvisualisasikan plot ini sebagai cummulative plot, juga dikenal sebagai __Stacked Line Plot__atau __Area Plot__

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

# get the top 5 entries
df_top5= df_can.head()

# transpose the dataframe
df_top5 = df_top5[years].transpose()

df_top5.head()
```

Area plots ter-stacked secara default. Dan untuk menghasilkan stacked area plot, setiap kolom harus berupa semua nilai positif atau semua negatif (Nilai NaN akan ditetapkan secara default ke 0)

Untuk menghasilkan unstacked plot, berikan stacked = False.

```python
df_top5.index = df_top5.index.map(int) # lets change the index values of df_top5 to type integer for plotting

df_top5.plot (kind='area',
              stacked=False,
              figsize=(20,10), # pass a tuple (x,y) size
             )

plt.title("Immigration Trend of Top 5 Countries")
plt.ylabel("Number of Immigrants")
plt.xlabel("Years")

plt.show()
```

Plot unstacked memiliki transparansi default (alpha value)) sebesar 0,5.
Kita dapat mengubah nilai ini dengan memasukkan parameter alpha.

```python
df_top5.plot(kind='area',
            alpha = 0.25,
            stacked=False,
            figsize=(20,10),)


plt.title("Immigration Trend of Top 5 Countries")
plt.ylabel("Number of Immigrants")
plt.xlabel("Years")

plt.show()
```

# Two types of plotting

Dua gara membuat plot dengan matplotlib. Membuat plot menggunakan Artist layer dan membuat plot menggunakan scripting layer.

__Option 1: Scripting layer (procedural method)__ - 
using matplotlib.pyplot as 'plt'

Kita dapat menggunakan plt yaitu maplotlib.pyplot dan menambahkan lebih banyak elemen dengan memanggil metode yang berbeda secara prosedural; misal plt.title atau plt.xlabel

```python
#Option 1: This is what we have been using so far 
df_top5.plot(kind='area', alpha=0.35, figsize=(20,10))

plt.title("Immigration Trend of Top 5 Countries")
plt.ylabel("Number of Immigrants")
plt.xlabel("Years")

plt.show()
```

__Option 2: Artist layer (Object oriented method)__ using an Axes instance from Matplotlib (preferred)

Kita dapat menggunakan instance Axes dari plot dan menyimpan dalam variabel (misal ax). kita dapat menambahkan lebih banyak elemen dengan memanggil metode dengan sedikit perubahann dalam sintaks (dgn menambahkan "set_" ke metode sebelumnya). misal: gunakan ax.set_title() sebagai ganti plt.title()

Di sesi ini menggunakan __scripting layer__, kecuali u/ beberapa visualisasi lanjutan dimana kita perlu menggunakan __artist layer__ u/ manipluasi aspek lanjutan dari plot.

```python
# Option 2 : preferred option with more flexibility 
ax = df_top5.plot(kind='area', alpha=0.35, figsize=(20,10))

plt.title("Immigration Trend of Top 5 Countries")
plt.ylabel("Number of Immigrants")
plt.xlabel("Years")
```

Mari buat beberapa modifikasi u/ meningkatkan impact dan estetika
1. increase the bin size to 15 by passing in bins parameter
2. set transparency to 60% by passing in alpha parameter
3. label the x-axis by passing in x-label paramter
4. Change the colors of the plots by passing in color parameter



1. tingkatkan ukuran bin menjadi 15 dengan memasukkan parameter bins
2. atur transparansi menjadi 60% dengan meneruskan parameter alfa
3. beri label sumbu-x dengan melewatkan parameter label-x
4. Ubah warna plot dengan memasukkan parameter warna 

```python
# lets get the x-tick values
count, bin_edges = np.histogram(df_t, 15)

# un-stacked histogram
df_t.plot(kind ='hist',
         figsize = (10,6),
         bins=15,
         alpha=0.6,
         xticks = bin_edges,
         color=['coral', 'darkslateblue','mediumseagreen']
         )
plt.title('Histogram of Immigration from Denmark, Norway, adn Sweden from 1980-2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

Jika tidak ingin plot saling tumpang tindih, kita dapat menumpuknya dengan parameter __stacaked__.
Menyelesaikan label sumbu x min dn maks u/ menghilangkan celah ekstra di tepi plot. kita bisa pass tuple (min, max) menggunakan parameter __xlim__

```python
count, bin_edges = np.histogram (df_t,15)
xmin = bin_edges[0] - 10 # first bin value is 31.10, adding buffer of 10 for aesthetic purposes
xmax = bin_edges[-1] + 10 # last bin value is 308.0, adding buffer of 10 for aesthetic purposes

# stacked Histogram
df_t.plot(kind='hist',
          figsize=(10,6),
          bins=15,
          xticks=bin_edges,
          color=['coral', 'darkslateblue','mediumseagreen'],
          stacked=True,
          xlim=(xmin, xmax)
        )
plt.title('Histogram of Immigration from Denmark, Norway, adn Sweden from 1980-2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

Mari beri anotasi pada plot menggunakan metode annotate pada scripting layer atau antarmuka pyplot. kita akan passing parameter berikut:
1. str : str, teks anotasi 
2. xy : Tuple yang menentukan titik (x,y) untuk diberi anotasi (in this case, end point of arrow)
3. xytext : Tuple yang menentukan titik (x,y) untuk menempatkan teks (in this case, start point of arrow)
4. xycoords : Sistem koordinat yang diberikan xy dalam -'data' menggunakan sistem koordinat objek yang dianotasi (default)
5. arrowprops : takes a dictionary of propersties to draw the arrow:
    - arrowstyle : specifies the arrow style, '->' is standard arrow
    - connectionstyle : specifies the connection type. arc3 is a straight line
    - color : specifies color of arror
    - lw : specifies the line witdh

Kita akan meneruskan parameter tambahan berikut:
- rotation : rotation angle of text in degress (counter clockwise)
- va : vertical aligment of the text [center, top, bottom, baseline]
- ha : horizontal alignment of text [center, right, left]

```python
df_iceland.plot(kind='bar', figsize=(10,6), rot=90)

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Iceland Immigrants to canada from 1980 to 2013')


# Annotate arrow
plt.annotate('', # s:str. will leave it blank for no text
            xy =(32,70), #tempatkan kepala panah pada titik (tahun 2012, pop 70) 
            xytext=(28,20), # tempatkan pangkal panah pada titik (tahun 2208, pop 20) 
            xycoords='data', # akan menggunakan sistem koordinat objek yang dianotasi 
            arrowprops = dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
            )

# Annotate Text
plt.annotate('2008-2011 Financial Crisis', # text to display
             xy=(28,30), # start the next at at point (year 2008, pop 30)
             rotation=72.5, # based on trial and eror to match
             va ='bottom', # want the text to be vertically 'bottom' aligned
            )
plt.show()
```

# __Pie Charts__

Membuat pie chart di Matplotlib dengan memasukkan kind=pie

Mari gunakan pie chart untuk menjelajahi proporsi imigran baru yang dikelompokkan berdasarkan benua untuk seluruh periode waktu dari 1980 hingga 2013.

Step 1 : Gather data

Kita akan menggunakan metode pandas groupby u/ meringkas data imigrasi berdasrkan Continent. Proses umum groupby melibatkan langkah berikut:
1. Split : membagi data menjadi beberapa kelompok berdasarkan beberapa kriteria.

2. Apply : Menerapkan fungsi ke setiap grup secara independen:.sum() .count() .mean() .std() .aggregate() .apply() .etc.

3. Combine : Menggabungkan hasil menjadi struktur data 

```python
# group contries by continents and apply sum() function
df_continents = df_can.groupby('Continent', axis=0).sum()

# note : the output of the groupby method is a groupby object
# we can not use it further until we apply a function (eg .sum())
print(type(df_can.groupby('Continent', axis=0)))

df_continents.head()
```

__Step 2: plot datanya__

Kita akan passing kind='pie', bersama parameter tambahan berikut:

1. autopct : adalah string atau fungsi yang digunakan untuk memberi label pada irisan dengan nilai numeriknya. Label akan ditempatkan di dalam irisan. jika itu adalah format string, labelnya akan bi fmt%pct
2. stratangle : memutar awal diagram lingkaran dengan sudut derajat berlawanan arah jarum jam dari sumbu x
3. shadow : menggambar bayangan di bawah pie (untuk memberi kesan 3d)

```python
# autopct create %, strat angle represent starting point 
df_continents['Total'].plot(kind='pie',
                           figsize=(5,6),
                           autopct='%1.1f%%', # add in precentages
                           startangle=90, #start angle 90o (Africa)
                           shadow=True, #add shadow
                           )
plt.title('Immigration to Canada by Continent [1980-2013]')
plt.axis('equal') #sets the pie chart look like a circle

plt.show()
```

Visual diatas tidak begitu jelas. bari kita buat modifikasi:

- Hapus label teks pada diagram lingkaran dengan memasukkan __legend__ dan tambahkan sebagai legenda terpisah menggunakan __plt.legend()__
- Dorong persentase untuk duduk tepat di luar diagram lingkaran dengan memasukkan parameter __pctdistance__
- Berikan set khusus __colors__ untuk benua dengan memasukkan parameter warna
- Explode diagram lingkaran untuk menekankan tiga benua terendah (Afrika, Amerika Utara, dan Amerika Latin dan Karibia) dengan memasukkan parameter __explode__.

```python
colors_list =['gold','yellowgreen','lightcoral','lightskyblue','lightgreen','pink']
explode_list =[0.1, 0, 0, 0, 0.1, 0.1] #rasio untuk setiap benua untuk mengimbangi setiap irisan

df_continents['Total'].plot(kind='pie',
                           figsize=(15,6),
                           autopct='%1.1f%%',
                            startangle=90,
                            shadow=True,
                            labels=None, # trun off labels on pie chart
                            pctdistance=1.12, # rasio antara bagian tengah setiap irisan pie dan tempat teks yang dihasilkan oleh autopct 
                            colors=colors_list, # add custom colors
                            explode= explode_list #explode lowest 3 continents
                           )

# scale the title up by 12% to match pctdistance
plt.title('Immigration to Canada by Continent [1980-2013]', y=1.12)
plt.axis('equal')

# add legend
plt.legend(labels=df_continents.index, loc='upper left')
plt.show()
```

# Bubble Plots (Optional)

Bubble plot adalah variasi scatter plot yang menampilkan tiga dimensi data (x,y,z).
titik data diganti dengan bubbles, dengan ukuran bubbles ditentukan oleh variabel ketiga 'z',/ bobot/

__Lets start analyzing the effect of Argentina's great depression__

Argentina mengalami great depression dari tahun 1998-2002, yang menyebabkan meluasnya pengangguran, kerusuhan, jatuhnya pemerintah, dan gagal bayar hutang luar negeri. Dalam hal pendapatan, lebih dari 50% orang argentina miskin dan tujuh dari sepulurh anak argentina miskin pada saat krisis tahun 2002.

Mari kita analisis dampak krisis ini dan bandingkan imigrasi argentina dengan tetangganya brasil. Mari kita lakukan itu dengan bubble plot imigrais dari brasil dan argentina untuk tahun 1980-2013. Kita akan menetapkan bobot gelembung sebagai nilai populasi yang dinormalisasi untuk setiap tahun

Step 1: Get the data for Brazil dan Argentina. We will convert the years to type int and bring it in the dataframe.

```python
df_can_t = df_can[years].transpose() #transposed dataframe

# cast the years (the index) to type int
df_can_t.index = map(int, df_can_t.index)

# lets label the index. this will automatically be the column name when we reset the index
df_can_t.index.name='Year'

# reset index to bring the Yaear in as a column
df_can_t.reset_index(inplace=True)

# view the canghes
df_can_t.head()
```

Step 2 : Create normalized wights

Ada beberapa metode normalisasi dalam statistik. Dalam kasus ini akan menggunakan feature scaling u/ membawa semua nilai ke dalam kisaran [0,1]

```python
# normalize Brazil data
norm_brazil = (df_can_t['Brazil'] - df_can_t['Brazil'].min()) /(df_can_t['Brazil'].max() -  df_can_t['Brazil'].min())

norm_argentina = (df_can_t['Argentina'] - df_can_t['Argentina'].min() )/ (df_can_t['Argentina'].max() - df_can_t['Argentina'].min())
```

Step 3 : Plot the data

- To plot two different scatter plots in one plot, kita dapat memasukkan sumbu satu plot ke plot lainnya tetapi meneruskannya melalui parameter ax
- kami juga akan meneruskan bobot menggunakan parameter s. mengingat bahwa bobot yang dinormalisasi adalah antara 0-1, bobot tersebut tidak akan terlihat di plot. Oleh karena itu kami akan : 
    1. kalikan bobot dengan 2000 untuk menaikkan skala grafik 
    2. tambahkan 10 untuk mengkompensasi nilai min (yang memiliki bobot 0 dan karenanya skala dengan x2000)

```python
#Brazil
ax0 = df_can_t.plot(kind='scatter',
                    x = 'Year',
                    y = 'Brazil',
                    figsize =(14,8),
                    alpha = 0.5, #tranparan
                    color = 'green',
                    s = norm_brazil * 2000 + 10, # pass in weights
                    xlim = (1975, 2015)
                    )
#Argentina
ax1 = df_can_t.plot(kind='scatter',
                    x = 'Year',
                    y = 'Argentina',
                    figsize =(14,8),
                    alpha = 0.5, #tranparan
                    color = 'blue',
                    s = norm_argentina * 2000 + 10, # pass in weights
                    ax = ax0
                    )

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration form Brazil and Argentina from 1980-2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```
