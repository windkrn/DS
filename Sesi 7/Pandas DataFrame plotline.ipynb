# Pandas Data Frame plot.line

DataFrame.plot.line(x=None, y=None, **kwargs)

Parameters
1. xlabel or position, optional
Allows plotting of one column versus another. If not specified, the index of the DataFrame is used.

2. ylabel or position, optional
Allows plotting of one column versus another. If not specified, all numerical columns are used.

3. colorstr, array-like, or dict, optional
The color for each of the DataFrame’s columns. Possible values are:

- A single color string referred to by name, RGB or RGBA code,
  for instance ‘red’ or ‘#a98d19’.

- A sequence of color strings referred to by name, RGB or RGBA
  code, which will be used for each column recursively. For instance [‘green’,’yellow’] each column’s line will be filled in 
  green or yellow, alternatively. If there is only a single column to be plotted, then only the first color from the color list 
  will be used.

- A dict of the form {column namecolor}, so that each column will be
  colored accordingly. For example, if your columns are called a and b, then passing {‘a’: ‘green’, ‘b’: ‘red’} will color 
  lines for column a in green and lines for column b in red.

New in version 1.1.0.

**kwargs
Additional keyword arguments are documented in DataFrame.plot().

Returns
matplotlib.axes.Axes or np.ndarray of them
An ndarray is returned with one matplotlib.axes.Axes per column when subplots=True.

```python
import pandas as pd
s = pd.Series([1, 3, 2])
s.plot.line()
```

```python
df = pd.DataFrame({
    'pig':[20, 18, 498, 675, 1776],
    'hourse':[4, 25, 281, 600, 1900]
}, index =[1990, 1997, 2003, 2009, 2014])
lines=df.plot.line()
```

```python
axes= df.plot.line(subplots=True)
type(axes)
```

```python
axes = df.plot.line(
    subplots=True, color={'pig':'pink', 'hourse':'#742802'}
)
```

```python
lines= df.plot.line(x='pig',y='hourse')
```
